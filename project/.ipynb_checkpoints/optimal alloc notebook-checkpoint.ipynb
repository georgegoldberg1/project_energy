```python
def get_optimal_allocation(models, min_budget, max_budget, total_budget, budget_bounds, min_media_budget_dict={}):
    global medias

    # Define min_budget depending on sliders or forced min_budgets (tweaks)
    for media in medias:
        min_budget[media] = int(min(max(min_budget[media], min_media_budget_dict.get(media, 0)), max_budget[media]))

    # Avoid bugs by setting the max to at least 1 when the sliders are set to 0
    for media in medias:
        max_budget[media] = max(1, max_budget[media])

    # If we min and max media budgets are equal it create bugs
    for media in medias:
        if max_budget[media] == min_budget[media]:
            max_budget[media] = min_budget[media] + 1

    # Define boundaries and media budgets for the optimizer
    bnds = []
    for media in medias:
        bnds.append((float(min_budget[media]), float(max_budget[media])))

    # Create default media budgets array
    media_budgets = [i[-1] for i in bnds]

    # Function that returns the revenue generated by a media for a specific budget
    def get_return(media, budget):
        std = self.media_params[media]['std']
        alpha = self.media_params[media]['alpha']
        beta = self.media_params[media]['beta']
        mu = self.media_params[media]['mu']

        # For each specific budgets we want spread this budget over a year to take into account adstock
##NO- DO NOT USE THIS-->         spend = np.repeat(budget/std, 52) #NO- DO NOT USE THIS
        spend = np.array([budget/std, *np.zeros(11)]) #<- USE THIS INSTEAD. 
    #explanation: the model has done adstock properly, so you must do it properly here too 
    #np.repeat is not good as it doesnt let the effect decay, inflating the budgets on all adstock channels 
    #AND THEREFORE ALL CHANNELS (as almost all are adstock ones.) returns willl seem very high because you 
    #are increasing spend by orders of magnitude in the eyes of the model. BUT ALSO NEED TO DROP THE /52 afterwards??
    
        adstock_output = geometric_adstock(spend, alpha=alpha)
        logistic = logistic_function(adstock_output, mu)
        output = self.kpi_std * logistic * beta
        return sum(output)
        #DONT DO THE BELOW IF YOU REMOVE Np.REPEATS?
#         return sum(output) / 52

    # Define objective function
    def objective(media_budgets, sign=-1.0):
        total_return = 0
        # Loop over each media 
        for index, media_budget in enumerate(media_budgets):
            if self.kpi_is_revenue:
                # Don't spend more if it doesn't increase the expected revenue by at least the same amount
                total_return += (get_return(medias[index], budget=media_budget) - media_budget)
            else:
                # Don't spend more if it doesn't increase the expected kpi
                total_return += get_return(medias[index], budget=media_budget)
        return (sign*total_return)

    # Define total budget constraint
    def constraint_total_get(media_budgets):
        return total_budget - sum(media_budgets)

    # Define max total budget constraint
    con1 = {'type': 'ineq', 'fun': constraint_total_get}
    cons = [con1]

    # Solve the optimization
    sol = minimize(objective, media_budgets, method='SLSQP', constraints=cons, bounds=bnds)
    self.expected_revenue = int(-sol['fun'])
    sol = [int(round(i,0)) for i in sol['x']]

    # If we optimize the profit we will get the profit so we need to add the total spend by media in order to get the revenue
    self.expected_revenue = self.expected_revenue if not self.kpi_is_revenue else self.expected_revenue + sum(sol)

    # Create dictionnary media:optimal_budget
    optimal_budgets = {[*min_budget.keys()][k]:sol[k] for k, v in enumerate(sol)}
    optimal_budgets_with_revenue = {media:{'budget': budget, 'revenue': get_return(budget=budget, media=media)} for media, budget in optimal_budgets.items()}

    # Result DataFrame
    self.df = pd.DataFrame([optimal_budgets]).T.rename(columns={0:"Media Spend in Â£"})
    self.df.index.name = "Media"
    x_axis_max_budget = int(self.df.max() * 1.4) # This is gonna be the x_axis max budget for the media curves

    # Create dictionary containing DataFrame with media_name, budget and corresponding revenue
    dfs_curves = {}
    for media in min_budget.keys():
        df_curves = pd.DataFrame()
        df_curves['Budget'] = np.linspace(0, x_axis_max_budget, 501) # Use limit to x_axis_max_budget instead of total_budget
        df_curves[media] = df_curves.Budget.map(lambda x: get_return(budget=x, media=media))
        dfs_curves[media] = df_curves

    # Add Scenario to Scenarios DataFrame
    tmp_df = self.df.T.copy()
    tmp_df.columns.name = ""
    tmp_df.reset_index(drop=True, inplace=True)
    tmp_df = tmp_df.rename(index={0: f'#{len(self.scenario_df) + 1}'})
    tmp_df.index.name = "Scenario"
    tmp_df["Budget"] = round_ten(tmp_df[self.media_cols].sum().sum())
    tmp_df[self.kpi_name] = int(self.expected_revenue)
    tmp_df["ROAS"] = round(self.expected_revenue / tmp_df["Budget"], 2)
    for col in self.media_cols:
        tmp_df[col] = round_ten(tmp_df[col])

    # Set the media curves and the cumulative curves in the app
    self.media_curves, self.cumulative_curve = optimal_curves(dfs_curves, total_budget, models, optimal_budgets_with_revenue)

    return tmp_df
```
